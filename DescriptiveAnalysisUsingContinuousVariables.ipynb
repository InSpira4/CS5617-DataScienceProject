```python
def scatterPlotCPTvsVar(plotVar) :
    #Scatterplot of CPT and plotVar, with the color of each point determined by accepted status
    figure(num=None, figsize=(30, 6), dpi=80, facecolor='w', edgecolor='k')
    groups = df.groupby("accepted")
    for name, group in reversed(tuple(groups)):
        if(name==False):
            plt.plot(group["CPT"], group[plotVar], color='r', alpha=.3, marker="o", linestyle="", label="denied")
        if(name==True):
            plt.plot(group["CPT"], group[plotVar], color='g', alpha=.3, marker="o", linestyle="", label="accepted")
    plt.legend()

    plt.xlabel("CPT")
    plt.ylabel(plotVar)
    plt.title("Scatterplot of CPT and "+ plotVar +" with the color of each point determined by accepted status")

```

```python
#Scatterplot of CPT and Charges, indicating acceptance status by color
scatterPlotCPTvsVar("Charges")
```

all the charges for 99999 got accepted and other accepted vs denied shows nearly same distribution over charges, it shows as denied because of overlaps

```python
def scatterPlotCPTSubPlots(plotVar) :
    #Scatterplot of CPT and plotVar, with the color of each point determined by accepted status
#     figure(num=None, figsize=(30, 6), dpi=80, facecolor='w', edgecolor='k')
    fig, axs = plt.subplots(2,1, figsize=(30, 6), facecolor='w', edgecolor='k')
    fig.subplots_adjust(hspace = .5, wspace=.001)

    axs = axs.ravel()
    groups = df.groupby("accepted")
    for name, group in reversed(tuple(groups)):
        if(name==False):
            axs[0].plot(group["CPT"], group[plotVar], color='r', alpha=.3, marker="o", linestyle="", label="denied")
#             axs[2].plot(group["CPT"], group[plotVar], color='r', alpha=.3, marker="o", linestyle="", label="denied")
        if(name==True):
            axs[1].plot(group["CPT"], group[plotVar], color='g', alpha=.3, marker="o", linestyle="", label="accepted")
#             axs[2].plot(group["CPT"], group[plotVar], color='g', alpha=.3, marker="o", linestyle="", label="accepted")
            
    for i in range(2):
        axs[i].legend()
    
        axs[i].set_xlabel("CPT")
        axs[i].set_ylabel(plotVar)

scatterPlotCPTSubPlots("Charges")   
```

```python
#Scatterplot of CPT and Payments, indicating acceptance status by color
scatterPlotCPTvsVar("Payments")
scatterPlotCPTSubPlots("Payments") 
```
