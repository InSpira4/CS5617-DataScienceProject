```python
def scatterPlotCPTvsVar(plotVar) :
    #Scatterplot of CPT and plotVar, with the color of each point determined by accepted status
    figure(num=None, figsize=(30, 6), dpi=80, facecolor='w', edgecolor='k')
    groups = df.groupby("accepted")
    for name, group in reversed(tuple(groups)):
        if(name==False):
            plt.plot(group["CPT"], group[plotVar], color='r', alpha=.3, marker="o", linestyle="", label="rejected")
        if(name==True):
            plt.plot(group["CPT"], group[plotVar], color='g', alpha=.3, marker="o", linestyle="", label="accepted")
    plt.legend()

    plt.xlabel("CPT")
    plt.ylabel(plotVar)
    plt.title("Scatterplot of CPT and "+ plotVar +" with the color of each point determined by accepted status")

```

```python
#Scatterplot of CPT and Charges, indicating acceptance status by color
scatterPlotCPTvsVar("Charges")

#Scatterplot of CPT and Payments, indicating acceptance status by color
scatterPlotCPTvsVar("Payments")

#Scatterplot of CPT and InsurancePayment, indicating acceptance status by color
scatterPlotCPTvsVar("InsurancePayment")

#Scatterplot of CPT and PatientPayment, indicating acceptance status by color
scatterPlotCPTvsVar("PatientPayment")
```

```python
#Scatterplot of CPT and Writeoffs, indicating acceptance status by color
scatterPlotCPTvsVar("Writeoffs")

#Scatterplot of CPT and InsuranceBal, indicating acceptance status by color
scatterPlotCPTvsVar("InsuranceBal")

#Scatterplot of CPT and PatientBal, indicating acceptance status by color
scatterPlotCPTvsVar("PatientBal")

#Scatterplot of CPT and AllowedAmount, indicating acceptance status by color
scatterPlotCPTvsVar("AllowedAmount")
```
